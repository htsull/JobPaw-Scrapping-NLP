```python
from bs4 import BeautifulSoup as bs
import requests
```

```python
r = requests.get("https://www.jobpaw.com/pont/professionnels.php?id=55")

# Convert to a beautiful soup object
soup = bs(r.text, "html.parser")

# Print out the HTML
contents = soup.prettify()
print(contents)
```

```python
jobs = soup.select(".table-hover td a")
jobs[0]['href']
```

```python
len(jobs)
```

```python
all = []
for job in jobs :
    all.append('https://www.jobpaw.com/pont/professionnels.php'+job['href'])
    # print('https://www.jobpaw.com/pont/professionnels.php'+job['href'])

a = [x for x in list(set(all)) if 'id=55' not in x]
a
```

```python
jobTable = soup.find(class_ = "table table-hover table-bordered")
tableRows = jobTable.find_all("tr")
for row in tableRows:
    print(row.prettify())
```
